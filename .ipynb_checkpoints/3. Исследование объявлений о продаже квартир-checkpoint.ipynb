## Предобработка данных

**Определим и изучим пропущенные значения.**

```python
data.isna().sum()
```

**Проверим какую долю составляют пропуски в каждом столбце.**

```python
data.isna().sum()/len(data)*100
```

```python
data.isna().sum().plot(kind='bar', figsize=(10, 4))
plt.show()
```

* В таблице много пропусков. Начнём заполнять подходящими значениями.

* Пропущенные значения в столбце `balcony` заменим на 0, если человек не указал число балконов - скорее всего, их нет. Приведём столбец к типу int64.

```python
data['balcony'] = data['balcony'].fillna(0)
data['balcony'] = data['balcony'].astype('int64')
data['balcony'].isna().sum()
```

**Проверим какие значения есть в столбце `ceiling_height` - высота потолков.**

```python
data['ceiling_height'].unique()
```

* Есть пропущенные значения *nan* и так же аномально низкие и высокие значения 1, 32, 26, 100 и другие. Максимальная высота потолков в Ленинградской области составляет 5.1 м, минимальная не меньше 2.5 м, все значения выше и ниже будем считать аномальными.


**Напишем функцию для замены аномальных значений.**

```python
def ceiling_height(row):    
    if row['ceiling_height'] < 2.5:
        return 2.5
    if row['ceiling_height'] > 5.1:
        return 5.1
    else:
        return row['ceiling_height']
data['ceiling_height'] = data.apply(ceiling_height, axis=1)
data['ceiling_height'].unique()
```

* Пропущенные значения оставим в таблице, предполагаю, что некоторые владельцы квартир не считают этот параметр обязательным или не знают высоту потолков, и поэтому не предприняли никаких действий для того, чтоб узнать и указать высоту потолков. Нам эти данные в дальнейшем исследовании не понадобятся, будем работать с тем что есть.

* В столбце `days_exposition` - количество дней размещения объявления (от публикации до снятия). Пропущенные значения оставим как есть, это могут быть непроданные квартиры, и мы не сможем рассчитать и заполнить корректно этот признак.

* В столбце `floors_total` - количество этажей 86 пропусков, ~ 0,3% от общей выборки, так же оставим пропуски в таблице.

* Далее заменим пропущенные значения в столбце `living_area` - общая жилая площадь, на медианное значение по столбцу с группировкой по количеству комнат.

```python
data['living_area'] = data['living_area'].fillna(data.groupby('rooms')['living_area'].transform("median"))
data['living_area'].isna().sum()
```

**Теперь заменим пропущенные значения в столбце `is_apartament`, если значение пропущено, заполним значением False. Изменим тип столбца на *bool*.**

```python
data['is_apartment'] = data['is_apartment'].fillna('False')
data['is_apartment'] = data['is_apartment'].astype('bool')
data['is_apartment'].isna().sum()
```

**В столбце `locality_name` избавимся от дубликатов, в словах 'посёлок' все буквы 'ё' заменим на 'е' и объединим поселок Мурино и Мурино. Так же приведём значения столбца к нижнему регистру.
Пропуски в столбце удалим.**

```python
data['locality_name'] = data['locality_name'].str.replace('ё', 'е')
data['locality_name'] = data['locality_name'].replace('поселок Мурино', 'Мурино')
data['locality_name'] = data['locality_name'].str.lower()
```

```python
data = data.dropna(subset=['locality_name'])
data['locality_name'].isna().sum()
```

**В столбце `city centers nearest` - расстояние до центра города, есть пропущенные значения, к сожалению корректно рассчитать этот признак мы не сможем, поэтому оставим пропуски в таблице.**

**Пропущенные значения площади кухни, где studio == True на 0.**

```python
data.loc[(data['studio'] == True) & (data['kitchen_area'].isna()), 'kitchen_area'] = 0
data['kitchen_area'].isna().sum()       
```

**Для оставшихся пропусков напишем функцию, которая заменит пропущенные значения на результат разницы между общей и жилой площадью.**

```python
def kitchen_area(row):
    living_area = row['living_area']
    total_area = row['total_area']
    kitchen_area = row['kitchen_area']
    if pd.isna(kitchen_area):
        return (total_area - living_area)
    else:
        return kitchen_area    
    
data['kitchen_area'] = data.apply(kitchen_area, axis=1)
data['kitchen_area'] = abs(data['kitchen_area'])
data['kitchen_area'].isna().sum()
```

* Пропусков в столбце `kitchen_area` больше нет, данные, которыми были заменены пропуски в `kitchen_area` не точные, так как ещё существуют прихожие, ванные комнаты, кладовки, балконы и т.п., а эти помещения так же как и кухня не входят в жилую площадь.

* Пропущенные данные в столбцах `airport nearest` - расстояние до аэропорта, `park around 3000` - кол-во парков в радиусе 3 км, `park nearest` - расстояние до ближайшего парка, `ponds around 3000` - кол-во водоёмов в радиусе 3 км, `ponds nearest` -  расстояние до ближайшего водоёма оставим без изменений, для исследования они нам не понадобятся.

**Переведём столбец `publication_date` в формат *datetime*.**

```python
data['publication_date'] = pd.to_datetime(data['publication_date'], format = '%Y-%m-%dT%H:%M:%S')
data.info()
```

**Проверим есь ли в таблице дубликаты.**

```python
data.duplicated().sum()
```

* Явных дубликатов нет, неявные дубликаты ранее были обработаны.

## Расчёты и добавление результатов в таблицу

**Посчитаем и добавим в таблицу:**
* цену квадратного метра;
* день недели, месяц и год публикации объявления;
* этаж квартиры; варианты — первый, последний, другой;
* соотношение жилой и общей площади, а также отношение площади кухни к общей.

```python
# цена квадратного метра
data['price_per_square_meter'] = round(data['last_price'] / data['total_area'], 0)
```

```python
# день недели публикации объявления о продаже квартиры
data['publication_day_of_the_week'] = data['publication_date'].dt.weekday
```

```python
# месяц публикации объявления
data['publication_month'] = data['publication_date'].dt.month
```

```python
# год публикации объявления
data['publication_year'] = data['publication_date'].dt.year
```

```python
# функция для определения этажа квартиры
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 'первый'
    elif floor == floors_total:
        return 'последний'
    else:
        return 'другой'

data['floor_category'] = data.apply(floor_category, axis = 1)    
```

```python
# соотношение жилой и общей площади
data['living_total_ratio'] = round(data['living_area'] / data['total_area'], 3)
```

```python
# отношение площади кухни к общей
data['kitchen_total_area'] = round(data['kitchen_area'] / data['total_area'], 3)
data.head()
```

* Все столбцы были успешно добавлены в таблицу.

**Выберем 10 населённых пунктов с наибольшим числом объявлений и посчитаем среднюю цену квадратного метра в этих населённых пунктах.**

**Так же выделим населённый пункт с самой низкой ценой за 1 м².**

```python
locality_pivot[locality_pivot['mean_price_m²'] == locality_pivot['mean_price_m²'].min()]
```

**Переведём значения столбца `last_price` в тысячные значения и переведём столбец в тип *int64*, с помощью метода array(), так как в столбце есть пропущенные значения *nan*.**

```python
data['distance_to_center'] = round(data['city_centers_nearest'] / 1000, 0)
data['distance_to_center'] = pd.array(data['distance_to_center'], dtype='Int64')
data_pivot_center = data.query('locality_name == "санкт-петербург"').pivot_table(index='distance_to_center', values='last_price')
data_pivot_center.plot(style='o-', grid=True, figsize=(15, 5), xlim=(0, 30));
```
